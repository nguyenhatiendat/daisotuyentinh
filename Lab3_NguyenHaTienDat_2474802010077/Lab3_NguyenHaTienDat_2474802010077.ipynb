- KHỞI TẠO MA TRẬN:

```python
from scipy import linalg
import numpy as np
a=np.array([1,2,3])
print("Vector a:\n",a)
#Chuyển b thành dạng ma trận
b=np.array([(1+9j,2j,3j),(4j,5j,6j)])
print("\nMa trận số phức b:\n",b)
c=np.array([[(0.5,1.5,10),(3,2,1)],[(6,5,4),(7,8,9)]])
print("\nMảng 3D c:\n",c)
A=np.asmatrix(np.random.random((2,2)))
print("\nMa trận A (random 2x2):\n",A)
B=np.asmatrix(b)
print("\nMa trận B (chuyển từ b):\n",B)
C=np.asmatrix(np.random.random((10,5)))
print("\nMa trận C (random 10x5):\n",C)
D=np.asmatrix([[4,3],[2,6]])
print("\nMa trận D (tự khai báo):\n",D)
#Tạo ma trận đường chéo. 3 là ma trận 3x3, k=1 là đường chéo nằm phía trên đường chéo chính (k = 0).
F=np.eye(3,k=1)
print("\nMa trận F (eye, k=1):\n",F)
```

```python
import numpy as np
def print_matrix(e):
    print("Ma trận giả định: E =",e.tolist(),"\n")
def mul(a,e,label):
    print(f"{label}:")
    for c in e.T:
        print(""," + ".join(f"{x}×{y}"for x,y in zip(a[0],c)),"=",sum(x*y for x,y in zip(a[0],c)))
    print()
    return a@e
e=np.array([[1,-2,2],[-1,1,3],[1,-1,-4]])
print_matrix(e)
raw=[([13,11,17],"KHO"),([1,0,11],"A_H"),([17,34,5],"O.C"),([0,6,26],"_DƯ"),([33,0,14],"~_L"),([12,9,34],"IÊ."),([25,35,35],"U??")]
m={f"KQ{i+1}":(np.array([v]),s)for i,(v,s) in enumerate(raw)}
res,text=[],""
for k,(a,s)in m.items():
    r=mul(a,e,k)
    res.append((k,[int(x)for x in r[0]]))
    text+=s.replace(" ","")
print("Tất cả kết quả mã hóa theo từng vector:")
for k,r in res:
    print(f"{k}:{r}")
print("\nChuỗi ký tự ban đầu:",text)
```

```python
import numpy as np
#kiểm tra ma trận có khả nghịch không
def inv(m):
    try:
        np.linalg.inv(m)
        return True
    except:
        return False
#chuyển chuỗi thành ma trận 3 cột (dựa trên mã ASCII)
def to_mt(s):
    a=[ord(c)for c in s]
    while len(a)%3:a.append(0)
    return np.array(a).reshape(-1,3)
#chuyển ma trận về chuỗi (bỏ số 0 dư)
def to_str(mt):
    return''.join(chr(int(round(x))) for r in mt for x in r if round(x))
def main():
    print("MÃ HÓA & GIẢI MÃ MA TRẬN 3x3\n")
    #ma trận khóa 3x3 (cần khả nghịch để giải mã)
    k=np.array([[2,-1,0],[1,1,1],[0,-1,2]])
    print(f"Khóa:\n{k}\n")
    if not inv(k):
        print("Không khả nghịch")
        return
    print("Khả nghịch\n")
    #nhập chuỗi (tên, MSSV, v.v.)
    s=input("Nhập họ tên hoặc MSSV:")
    #chuyển chuỗi thành ma trận
    a=to_mt(s)
    print(f"\nASCII:\n{a}")
    #nhân ma trận dữ liệu với khóa
    m=a@k
    print(f"\nMã hóa:\n{m}")
    #nhân với nghịch đảo của khóa
    d=m@np.linalg.inv(k)
    print(f"\nGiải mã:\n{d}")
    #in ra chuỗi đã giải mã
    print(f"\nKết quả: {to_str(d)}")
if __name__=="__main__":
    main()
```

```python
import numpy as np
#tính p/q theo công thức liên phân số
def tinh_pq(cs):
    p0,p1=1,cs[0]
    q0,q1=0,1
    for c in cs[1:]:
        p0,p1=p1,c*p1+p0
        q0,q1=q1,c*q1+q0
    return p1,q1
#kiểm tra kết quả bằng công thức ma trận
def kt_mt(cs):
    kq=np.identity(2,dtype=int)
    for c in cs:
        A=np.array([[c,1],[1,0]])
        kq=kq@A
    v=np.array([[1],[0]])
    return (kq@v).flatten()
def hien_thi(cs,lb):
    print(f"\n{lb}:")
    p,q=tinh_pq(cs)
    print(f"a. Tính tay: p={p}, q={q}->{p}/{q}")
    p2,q2=kt_mt(cs)
    print(f"b. Ma trận: p={p2}, q={q2}")
def main():
    print("\n CÂU 2: LIÊN PHÂN SỐ")
    hien_thi([2,1,2],"Phân số 8/3")
    hien_thi([3,1,2],"Phân số 11/3")
if __name__=="__main__":
    main()
```
